Currently there is a base model, in which the values for the seven adjusted parameters are

\begin{align}
    \bullet \quad \gamma = 2 \text{ (CRRA elasticity of substitution)} \\
    \bullet \quad \sigma_z = 0.0521 \text{ (SD of income process white noise)} \\
    \bullet \quad \alpha = 0.14 \text{ (1 - share of income for consumption)} \\
    \bullet \quad \theta = 0.25 \text{ (Down payment parameter)} \\
    \bullet \quad r_{rent} = .0755 \text{ (Rental rate (?))} \\
    \bullet \quad a_0 = 0 \text{ (Initial level of assets)} \\
    \bullet \quad D_0 = 0 \text{ (Initial level of durables)} \\
\end{align}

For $\gamma$, $\alpha$, $\theta$ and $r_{rent}$ there is a "low" model that sets the parameter value one percent lower than its base value, and a "high" one that sets it one percent higher. $\sigma_z$ includes those as well as one setting it to 0.

The settings for the assets and durables are a bit more complicated, as will be seen later.


Before showing the distributions, some summary statistics:

```python
allM = ['base', 'low_s', 'high_s', 'low_gamma', 'high_gamma', 'low_alpha', 'high_alpha', 
        'low_theta', 'high_theta', 'low_rrental', 'high_rrental', 'med_assets',
        'high_assets', 'med_durables', 'high_durables', 'no_s']
base.appendModels('fthb_dist', model=allM)
base.addLog()
baseMean = base.appended.set_index('model').loc['base', 'logAge'].mean()
baseScale = base.appended.set_index('model').loc['base', 'logAge'].std()
out = base.summaryTable('ageBought')
out
```

```python
# Test conformity with lognormal distribution
# (if t is considered as a stopping time)
figData = base.appended.unstack('model')
def xlabels(axes): # Scaling age axis for better interpretation
    axes.set_xticklabels([25 + 10*x for x in range(0, 7)])

pd.options.display.mpl_style = None
plt.figure(1)
actual = figData.plot(y=[('ageBought', 'base')], kind='kde', legend=False, xlim=(0,60))
xlabels(actual)
x = np.linspace(0,60, 200)
hypo = plt.plot(x, lognorm.pdf(x, baseScale, scale=np.exp(baseMean)))
plt.show()
```

The blue density is the kernel density of the age when simulated agents bought their first house.
The green density is a log normal fit with the same parameters that exists in the data
(i.e. SD of the logged variable and its mean, exponentiated)

```python
# Check determinism without individual shocks
noSigma = figData.plot(y=[('ageBought', 'no_s')], kind='hist', legend=False, xlim=(0,60))
xlabels(noSigma)
```

This is the distribution for $\sigma_z$ = 0. As also seen in the summary table, all values are 19.

* * *

The actual distributions are below. Blue is always the base model. The green model is the "high" model and the red model is the "low" one. The caption also notes the elasticity of mean age with respect to the parameter.

```python
rel = figData.plot(y=[('ageBought', 'base'), ('ageBought', 'high_s'), ('ageBought', 'low_s')],
                   kind='kde', alpha=0.6, legend=False, xlim=(10,30))
def xlabels(axes): # Zooming in, so overriding this
    axes.set_xticklabels([25 + 10 + 5*x for x in range(0, 6)])

xlabels(rel)
```

**Changes in $\sigma_z$.** High has elasticity 0.077, low has elasticity 0.046.

```python
rel = figData.plot(y=[('ageBought', 'base'), ('ageBought', 'high_gamma'), ('ageBought', 'low_gamma')],
                   kind='kde', alpha=0.6, legend=False, xlim=(10,30))
xlabels(rel)
```

**Changes in $\gamma$.** High has elasticity 0.159, low has elasticity -0.213.

```python
rel = figData.plot(y=[('ageBought', 'base'), ('ageBought', 'high_alpha'), ('ageBought', 'low_alpha')],
                   kind='kde', alpha=0.6, legend=False, xlim=(10,30))
xlabels(rel)
```

**Changes in $\alpha$.** High has elasticity 0.633. Low has elasticity -1.40.

```python
rel = figData.plot(y=[('ageBought', 'base'), ('ageBought', 'high_theta'), ('ageBought', 'low_theta')],
                   kind='kde', alpha=0.6, legend=False, xlim=(10,30))
xlabels(rel)
```

**Changes in $\theta$.** High has elasticity 0.45, low has elasticity -0.528.

```python
rel = figData.plot(y=[('ageBought', 'base'), ('ageBought', 'high_rrental'), ('ageBought', 'low_rrental')],
                   kind='kde', alpha=0.6, legend=False, xlim=(10,30))
xlabels(rel)
```

**Changes in $r_{rent}$.** High has elasticity -2.213, low has elasticity 1.379.

For initial assets and durables we do something different: show a table of mean values across many ($a_0, D_0$) specifications. Below, the x-axis gives the initial durables endowment for every agent and the Y gives the initial asset level. Data is plotted both as a table and as a heatmap, though notice the axes are inverted between them.

```python
from itertools import chain

initM = ['a%.1fD%.1f' % (x, y) for x in np.linspace(0, 0.6, 7) 
                                for y in np.linspace(0, 1.2, 7)]
base = lifecycle_iterate({'elasticity': 2}, baseFile)
base.appendModels('fthb_dist', model=initM)
base.appended['assets'] = base.appended['model'].str[:4]
base.appended['durables'] = base.appended['model'].str[4:]
base.appended.set_index(['id', 'assets', 'durables'], inplace=True)
out = base.appended.unstack('durables')['ageBought'].groupby(level=1).mean()
fig, ax = plt.subplots()
ax.pcolor(out, cmap=plt.cm.OrRd)

ax.set_xticks(np.arange(out.shape[0])+0.5, minor=False)
ax.set_yticks(np.arange(out.shape[0])+0.5, minor=False)

ax.set_xticklabels(np.linspace(0, 1.2, 7), minor=False)
ax.set_yticklabels(np.linspace(0, 0.6, 7), minor=False)
out
```
